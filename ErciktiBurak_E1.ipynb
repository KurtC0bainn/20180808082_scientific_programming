**֎Exercise 2.7 → C♣llatz sequence**



* x0 = 5 → sequence: 5 16 8 4 2 1
* C♣llatz sequence starting at x0 = 103.

```python
def collatz_sequence(n):
    print('▼Collatz sequence starting number is → {' + str(n) + '}▼')
    print(n, end = ' ')
    while n != 1:
        if n % 2 == 0:
            # If the # even, n ÷ 2
            n = int(n / 2)
        else:
            # If the # odd, triple it and add 1.
            n = 3 * n + 1
        print(n, end = ' ')
collatz_sequence(103)
```

**֎Exercise 3.4 → Primes:**

```python
from math import sqrt
def is_prime(n):
    if n < 2:
        return False
    # I iterate up to the square root of n + 1
    # If n is not prime → n = a * b → either a = b or a ≠ b → a < √n or b < √n 
    for i in range(2, int(sqrt(n) + 1)):
        if n % i == 0:
            return False
    return True

print(is_prime(25))
```

```python
from math import sqrt
import time

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(sqrt(n) + 1)):
        if n % i == 0:
            return False
    return True

def is_prime_new(n):
    if n < 2:
        return False
    # apart from 2 and 3
    if n == 2 or n == 3:
        return True
    # all primes are of the form 6k ± 1 → 6 * k - 1, 6 * k + 1 = (k - 1) % 6 != 0, (k + 1) % 6 != 0
    if (n + 1) % 6 != 0 and (n - 1) % 6 != 0:
        return False
    # I iterate up to the square root of n +1
    # If n is not prime → n = a * b → either a = b or a ≠ b → a < √n or b < √n 
    for i in range(2, int(sqrt(n) + 1)):
        if n % i == 0:
            return False
    return True

for N in range(37, 39):
    is_prime_start = time.perf_counter() * 1000
    prime = is_prime(N)
    is_prime_end = time.perf_counter() * 1000

    is_prime_start_new = time.perf_counter() * 1000
    prime_new = is_prime_new(N)
    is_prime_end_new = time.perf_counter() * 1000

    print()
    print(" (1) Prime number is:", N, "and running time is → ", is_prime_end - is_prime_start, " ms ")
    print(" (1) is prime?", str(prime))
    print(" (2) New Prime number is:", N, "and running time is → ", is_prime_end_new - is_prime_start_new, " ms ")
    print(" (2) is prime?", str(prime_new))
```
